# 2. Vizualizace dat

### Zkušenost zaměstnanců

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ds_salaries.csv')

df['experience_level'] = df['experience_level'].replace('EN','Junior')
df['experience_level'] = df['experience_level'].replace('MI','Medior')
df['experience_level'] = df['experience_level'].replace('SE','Senior')
df['experience_level'] = df['experience_level'].replace('EX','Executive-level')

exp_level = df['experience_level'].value_counts()
plt.pie(exp_level, labels=exp_level.index, autopct='%1.1f%%')
plt.title('Experience levels')
plt.axis('equal')
plt.show()
```

### Job titles

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ds_salaries.csv')

job_counts = df['job_title'].value_counts()

labels = job_counts.index.tolist()
sizes = job_counts.values.tolist()
total = sum(sizes)
percentages = [100 * size / total for size in sizes]
other_percentage = 100 - sum(percentages[:5])
other_label = 'Other'
percentages = percentages[:5] + [other_percentage]
labels = labels[:5] + [other_label]

plt.pie(percentages, labels=labels, autopct='%1.1f%%', startangle=90, counterclock=False)
plt.title('Job titles')
plt.show()

```

### Home office v čase

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ds_salaries.csv')

avg_remote_ratio_in_time = df.groupby('work_year')['remote_ratio'].mean().reset_index()

years = pd.unique(df.work_year)
plt.xticks(years)

plt.plot(avg_remote_ratio_in_time['work_year'], avg_remote_ratio_in_time['remote_ratio'])
plt.title('Average remote work ratio in time')
plt.xlabel('Year')
plt.ylabel('Average remote work ratio')
plt.show()
```

### Průměrná mzda podle zkušeností

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ds_salaries.csv')

df['experience_level'] = df['experience_level'].replace('EN','Junior')
df['experience_level'] = df['experience_level'].replace('MI','Medior')
df['experience_level'] = df['experience_level'].replace('SE','Senior')
df['experience_level'] = df['experience_level'].replace('EX','Executive-level')

avg_salary_by_exp = df.groupby('experience_level')['salary_in_usd'].mean().reset_index()
avg_salary_by_exp = avg_salary_by_exp.sort_values('salary_in_usd')

plt.bar(avg_salary_by_exp['experience_level'], avg_salary_by_exp['salary_in_usd'])
plt.title('Average salary by experience level')
plt.xlabel('Experience level')
plt.ylabel('Average salary (USD)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('ds_salaries.csv')

years = pd.unique(df.work_year)
plt.xticks(years)

avg_salary_in_time = df.groupby('work_year')['salary_in_usd'].mean().reset_index()
plt.plot(avg_salary_in_time['work_year'], avg_salary_in_time['salary_in_usd'])
plt.title('Average salary in time')
plt.xlabel('Year')
plt.ylabel('Average salary (USD)')
plt.show()
```
